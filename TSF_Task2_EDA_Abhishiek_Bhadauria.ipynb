```
#importing important libraries and reading dataset
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
dataset=pd.read_csv('/content/drive/MyDrive/SampleSuperstore.csv')
print(dataset.head)
dataset.shape
```

```
#Printing details of dataset
dataset.describe(include='all')
```

```
#Checking for null values
dataset.isnull().any()
```

```
#Data types present in dataset
dataset.dtypes
```

```
#Printing class count and plotting graph
class_counts=dataset.groupby('Ship Mode').size()
print(class_counts)
set1=dataset.groupby(['Ship Mode'],as_index=False)['Profit'].sum()
sns.barplot(set1['Ship Mode'],set1['Profit'])
```

```
class_counts=dataset.groupby('Segment').size()
print(class_counts)
set1=dataset.groupby(['Segment'],as_index=False)['Profit'].sum()
sns.barplot(set1['Segment'],set1['Profit'])
```

```
class_counts=dataset.groupby('Region').size()
print(class_counts)
set1=dataset.groupby(['Region'],as_index=False)['Profit'].sum()
sns.barplot(set1['Region'],set1['Profit'])
```

```
class_counts=dataset.groupby('Category').size()
print(class_counts)
set1=dataset.groupby(['Category'],as_index=False)['Profit'].sum()
sns.barplot(set1['Category'],set1['Profit'])
```

```
class_counts=dataset.groupby('Sub-Category').size()
print(class_counts)
plt.figure(figsize=(10,10))
dataset['Sub-Category'].value_counts().plot(kind='pie')
```

```
#A correlation matrix to see relationship between variables.
cr=dataset.corr()
plt.figure(figsize=(10,10))
sns.heatmap(cr,annot=True)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```
city=dataset['City'].value_counts()
city.plot(kind='line',figsize=(12,12))
plt.title("City Order")
plt.ylabel('Count')
plt.xlabel('City')
plt.show()
```

```
#sns pairplot
sns.pairplot(dataset,plot_kws=dict(alpha=.3,edgecolor='none'),height=2,aspect=1.1);
```
